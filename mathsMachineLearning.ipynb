```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define a range of parameter values
# For example, if you're plotting a cost function for a linear regression model with a single parameter (slope),
# you can define a range of possible slope values.
slope_values = np.linspace(-10, 10, 100)  # Adjust the range and number of points as needed

# Step 2: Define your cost function
# You'll need a cost function that calculates the cost or loss for each parameter value.
# This is a simplified example for a quadratic cost function.
def quadratic_cost(slope):
    # Define your model and data
    x = np.array([1, 2, 3, 4, 5])  # Example input data
    y = np.array([2, 4, 5, 4, 5])  # Example output data

    # Calculate the predicted values using the model
    predicted = slope * x

    # Calculate the mean squared error (MSE) as the cost
    mse = np.mean((predicted - y) ** 2)

    return mse

# Step 3: Calculate the cost for each parameter value
cost_values = [quadratic_cost(slope) for slope in slope_values]

# Step 4: Plot the cost function
plt.figure(figsize=(8, 6))
plt.plot(slope_values, cost_values, label='Cost Function')
plt.xlabel('Slope Parameter')
plt.ylabel('Cost')
plt.title('Cost Function Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# multi dimensional cost function
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Define a range of values for both parameters
# For simplicity, we'll use a grid of values for two parameters, param1 and param2.
param1_values = np.linspace(-10, 10, 100)  # Adjust the range and number of points as needed
param2_values = np.linspace(-10, 10, 100)  # Adjust the range and number of points as needed

# Create a grid of parameter combinations using NumPy's meshgrid
param1_grid, param2_grid = np.meshgrid(param1_values, param2_values)

# Step 2: Define your multidimensional cost function
# Replace this with your specific cost function that takes two parameters.
def multidimensional_cost(param1, param2):
    # Calculate the cost based on the parameters
    # This is a simplified example; replace it with your cost function.
    cost = param1 ** 2 + param2 ** 2
    return cost

# Calculate the cost for each parameter combination in the grid
cost_values = multidimensional_cost(param1_grid, param2_grid)

# Step 3: Create a contour or surface plot
fig = plt.figure(figsize=(10, 8))

# For a 2D contour plot:
plt.contour(param1_grid, param2_grid, cost_values, levels=20, cmap='viridis')
plt.xlabel('Parameter 1')
plt.ylabel('Parameter 2')
plt.title('Contour Plot of Multidimensional Cost Function')
plt.colorbar()

# For a 3D surface plot:
# ax = fig.add_subplot(111, projection='3d')
# ax.plot_surface(param1_grid, param2_grid, cost_values, cmap='viridis')
# ax.set_xlabel('Parameter 1')
# ax.set_ylabel('Parameter 2')
# ax.set_zlabel('Cost')
# ax.set_title('Surface Plot of Multidimensional Cost Function')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define a range of values for two parameters (param1 and param2)
param1_values = np.linspace(-10, 10, 100)  # Adjust the range and number of points as needed
param2_values = np.linspace(-10, 10, 100)  # Adjust the range and number of points as needed

# Create a grid of parameter combinations using NumPy's meshgrid
param1_grid, param2_grid = np.meshgrid(param1_values, param2_values)

# Define your 3D cost function
def cost_function(param1, param2):
    # Replace this with your specific cost function
    # For example, a simple quadratic cost function:
    cost = param1**2 + param2**2
    return cost

# Calculate the cost for each parameter combination in the grid
cost_values = cost_function(param1_grid, param2_grid)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(param1_grid, param2_grid, cost_values, cmap='viridis')
ax.set_xlabel('Parameter 1')
ax.set_ylabel('Parameter 2')
ax.set_zlabel('Cost')
ax.set_title('3D Surface Plot of Cost Function')

plt.show()
```
